```python
import streamlit as st
from ultralytics import YOLO
import cv2
from PIL import Image
import numpy as np

# Load trained model
model = YOLO("/content/drive/MyDrive/datasets/runs/detect/train2/weights/best.pt")  # apna best weight file ka naam daal

# ğŸ§­ Navbar / Title
st.set_page_config(page_title="Safety Detection App", page_icon="ğŸ¦º", layout="wide")
st.title("ğŸ¦º Safety Equipment Detection using YOLOv8")

# ğŸ“¸ Upload section
uploaded_file = st.file_uploader("Upload an image...", type=["jpg", "jpeg", "png"])

# ğŸ•¹ï¸ Detect button
if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="ğŸ“· Uploaded Image", use_column_width=True)

    if st.button("ğŸš€ Detect Objects"):
        img_array = np.array(image)
        results = model.predict(img_array, conf=0.25)

        # ğŸ“Š Convert result to image with bounding boxes
        annotated_frame = results[0].plot()
        st.image(annotated_frame, caption="âœ… Detection Result", use_column_width=True)

# ğŸ“ Footer
st.markdown("---")
st.markdown("ğŸ‘¨â€ğŸ’» *Developed by Team Codex for GenIgnite Hackathon 2025*")
```

```python
import gradio as gr
from ultralytics import YOLO
import cv2
import numpy as np

# Load trained model
# Make sure this path is correct based on your training output
model = YOLO("/content/runs/detect/train2/weights/best.pt") # Corrected model path

def detect_objects(image):
    """
    Performs object detection on an image using the loaded YOLO model.
    """
    if image is None:
        return None

    # Convert Gradio image (numpy array) to a format suitable for YOLO
    # Gradio provides image as a numpy array, which is directly usable by YOLO
    results = model.predict(image, conf=0.25)

    # Render results on the image
    annotated_frame = results[0].plot()

    return annotated_frame

# Create Gradio interface
iface = gr.Interface(
    fn=detect_objects,
    inputs=gr.Image(type="numpy", label="Upload Image or Use Webcam"), # Modified input type
    outputs=gr.Image(type="numpy", label="Detection Results"),
    title="ğŸ¦º Safety Equipment Detection using YOLOv8",
    description="Upload an image or use your webcam to detect safety equipment in real-time." # Modified description
)

# Launch the Gradio app
iface.launch(debug=True, share=True) # Added share=True for Colab
```
